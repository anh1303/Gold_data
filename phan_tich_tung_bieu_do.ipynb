# Phân tích tương quan chi tiết giá Vàng
Mỗi biểu đồ dưới đây phân tích mối quan hệ giữa giá đóng cửa (**Close**) và một thông số cụ thể.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

def clean_currency(x):
    if isinstance(x, str):
        x = x.replace(',', '')
        if 'K' in x: return float(x.replace('K', '')) * 1e3
        if 'M' in x: return float(x.replace('M', '')) * 1e6
    return pd.to_numeric(x, errors='coerce')

df = pd.read_csv('gold.csv')
for col in ['Close', 'Volume', 'Open', 'High', 'Low']:
    df[col] = df[col].apply(clean_currency)

df['Date'] = pd.to_datetime(df['Date'])
print('Dữ liệu đã được làm sạch và sẵn sàng.')
```

## Biểu đồ tương quan: Close vs Silver

```python
plt.figure(figsize=(10, 5))
corr = df['Close'].corr(df['Silver'])
sns.regplot(x='Silver', y='Close', data=df, scatter_kws={'alpha':0.3}, line_kws={'color':'red'})
plt.title(f'Tương quan Close và Silver (Pearson: {corr:.4f})')
plt.grid(True, alpha=0.3)
plt.show()
```
